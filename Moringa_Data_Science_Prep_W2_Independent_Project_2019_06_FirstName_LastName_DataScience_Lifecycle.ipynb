```python
%sql DROP TABLE combinedtable;
```

```python
# To compute the ratio between the number of grand electors and the population. Please create a new column with that ratio
# create new table
%%sql
CREATE TABLE combinedtable(State, Population, GrandElectors);
SELECT * FROM combinedtable
```

```python
# copy the data into the new table "combinedtable"
%%sql
INSERT INTO combinedtable SELECT population_by_state.State, Population, GrandElectors FROM population_by_state
INNER JOIN grand_electors_by_state ON LOWER(population_by_state.State) = LOWER(grand_electors_by_state.State);
SELECT * FROM combinedtable;
```

```python
# Add the new column ratio to combinedtable
%%sql
ALTER TABLE combinedtable ADD ratio REAL DEFAULT 'NULL';
SELECT * FROM combinedtable;
```

```python
# Update the column ratio
%%sql
UPDATE combinedtable 
SET ratio = GrandElectors/Population
WHERE ratio = "NULL";
SELECT * FROM combinedtable;
```

```python
# To order the states by decreasing ratio of Grand Electors per capital. That will make our priority list.
%%sql
SELECT * FROM combinedtable ORDER BY GrandElectors DESC;
```

```python
%sql DROP TABLE sortedcombineddata;
```

```python
# create new table "sortedcombineddata"
%%sql
CREATE TABLE sortedcombineddata(State, Population, GrandElectors, ratio);
SELECT * FROM sortedcombineddata;
```

```python
# copy the data into the new table "sortedcombineddata"
%%sql
INSERT INTO sortedcombineddata SELECT * FROM combinedtable ORDER BY  GrandElectors;
SELECT * FROM sortedcombineddata;
```

```python
# find the running sum
%%sql
SELECT a.State, a.Population, a.GrandElectors, a.ratio, SUM(b.GrandElectors) AS SUM
FROM sortedcombineddata AS a, sortedcombineddata AS b WHERE b.GrandElectors <= a.GrandElectors 
GROUP BY b.GrandElectors ORDER BY SUM(b.GrandElectors) DESC;
```

```python
# find the running halfsum
%%sql
SELECT a.State, a.Population, a.GrandElectors, a.ratio, SUM(b.GrandElectors)*1/2 AS HALFSUM
FROM sortedcombineddata AS a, sortedcombineddata AS b WHERE b.GrandElectors <= a.GrandElectors 
GROUP BY b.GrandElectors ORDER BY SUM(b.GrandElectors) DESC;
```

ACCORDING TO THE QUERY THE THRESHOLD SUM IS 630

```python
# To filter our sorted list of states in order to keep only the (top) ones enabling us to reach the computed threshold. (the other states can be ignored). That is our target list.
  # Hint: You can do that in 2 steps:
    # Select all the states for which the running total is below or equal to the threshold.
    # Add the first state for which the running total is larger than the threshold.
# CREATE NEW TABLE sortedcombineddata2
%%sql
CREATE TABLE sortedcombineddata2 (
    State,
    Population,
    GrandElectors,
    ratio,
    SUM
);

```

```python
%%sql
INSERT INTO sortedcombineddata2 SELECT a.State, a.Population, a.GrandElectors, a.ratio, SUM(b.GrandElectors) AS SUM
FROM sortedcombineddata AS a, sortedcombineddata AS b WHERE b.GrandElectors <= a.GrandElectors 
GROUP BY b.GrandElectors ORDER BY SUM(b.GrandElectors) DESC;
```

```python
%sql SELECT * FROM sortedcombineddata2;
```

```python
%%sql
SELECT * FROM sortedcombineddata2 WHERE SUM <= 630;
```
